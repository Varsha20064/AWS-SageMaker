```python
plt.figure(figsize=(8,5))

plt.bar(df["Name"], df["Salary"], color="skyblue")

plt.title("Employee Salary Comparison")

plt.xlabel("Employee Name")

plt.ylabel("Salary")

plt.show()
```

```python
dept = df["Department"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(dept,
        labels=dept.index,
        autopct="%1.1f%%",
        startangle=90)

plt.title("Department Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df["Name"],
         df["Experience"],
         marker="o")

plt.title("Employee Experience")

plt.xlabel("Employee Name")

plt.ylabel("Experience (Years)")

plt.grid(True)

plt.show()
```
